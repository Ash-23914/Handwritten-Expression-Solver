```python
import cv2
import os as os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
DATADIR = 'dataset'
CATEGORIES = ['0','1','3','4','5','6','7','8','9','add','dec','div','eq','mul','sub','x','y','z']
```

```python
IMG_SIZE = 100
training_data=[]
def create_training_data():
    for category in CATEGORIES:
        path=os.path.join(DATADIR, category)
        class_num=CATEGORIES.index(category)
        for filename in os.listdir(path):
            if filename != ".directory" :

                image = cv2.imread(os.path.join(path,filename),cv2.IMREAD_GRAYSCALE)
                image = ~image

                if image is not None:

                    ret, thresh = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
                    contours, hierarchy = cv2.findContours(thresh,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
                    contour = sorted(contours, key = lambda ctr: cv2.boundingRect(ctr)[0])

                    a = int(28)
                    b = int(28)
                    maxi = 0

                    for c in contour:

                        x,y,a,b=cv2.boundingRect(c)
                        maxi=max(a*b,maxi)

                        if maxi==a*b:

                            x_max=x
                            y_max=y
                            w_max=a
                            h_max=b

                    im_crop = thresh[y_max:y_max+h_max+10, x_max:x_max+w_max+10]
                    im_resize = cv2.resize(im_crop,(28,28))
                    im_resize = np.reshape(im_resize,(784,1))
                    training_data.append([im_resize,class_num])
                    print(class_num)
#             try:
#                 img_array=cv2.imread(os.path.join(path,img))
#                 print(img_array)
#                 new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
#                 training_data.append([new_array,class_num])
#             except Exception as e:
#                 pass
create_training_data()            
```

```python
X=[]
y=[]

for categories, label in training_data:
    X.append(categories)
    y.append(label)
X= np.array(X).reshape(lenofimage,-1)
##X = tf.keras.utils.normalize(X, axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
print(X_train)
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], -1))
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='rbf',gamma='auto')
svc.fit(X_train, y_train)
```

```python
y2 = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy on unknown data is",accuracy_score(y_test,y2))
```
